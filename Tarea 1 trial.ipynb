## cargar datos a dataframe pandas

```python
raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-2-4.json', lines=True)
print("loaded")
```

```python
raw_tweets = raw_tweets[raw_tweets['lang']=='en']
print("Shape: ", raw_tweets.shape)
```

```python
users = json_normalize(raw_tweets['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(5)
```

```python
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
print("Shape: ", users.shape)
users.head(5)
```

```python
user_id = []
for user in raw_tweets['user']:
    uid = user['id']
    user_id.append(uid)
raw_tweets['userId'] = user_id

# Remove less important columns
cols = ['url', 'date', 'renderedContent', 'id', 'userId', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'source', 'media', 'retweetedTweet', 'quotedTweet', 'mentionedUsers']
tweets = raw_tweets[cols]
tweets.rename(columns={'id':'tweetId', 'url':'tweetUrl'}, inplace=True)
tweets.head(5)
```

```python
tweets = pd.DataFrame(tweets)
tweets.drop_duplicates(subset=['tweetId'], inplace=True)
print("Shape: ", tweets.shape)
tweets.head(5)
```

## Escribo en ingles los enunciados siguientes por un tema de notación de código y no perder la consistencia

### top 10 most retweeted tweets

```python
def top_10_retweeted():
    newlist = tweets.nlargest(n=10, columns=['retweetCount'])
    display(newlist)
    return newlist
```

### top 10 users with most tweets

```python
def top_10_users_uploads():
    most_tweets = tweets["userId"].value_counts().reset_index()
    most_tweets.columns = ["userId", "count"]
    most_tweets.head(10)
    top_users = most_tweets.head(10)
    list_top_users = top_users["userId"].values.tolist()
    display(users[users['userId'].isin(list_top_users)])
    return users[users['userId'].isin(list_top_users)]
```

### Top 10 days with most tweets

#### First remove the time in the timestamp column by transforming it to only time type

```python
def top_10_days():
    tweets['dateNots'] = pd.to_datetime(tweets['date'])
    tweets['daydate'] = tweets['dateNots'].dt.date
    best_day = tweets["daydate"].value_counts().reset_index()
    best_day.columns = ["date", "tweet count"]
    top_days = best_day.head(10)
    list_top_days = top_days["date"].values.tolist()
    display(top_days)
    return top_days
```

### get the top 10 most used hashtags

```python
def top_10_hashtags():
    regular_expresion = "#\S+"
    hashtags=tweets['renderedContent'].str.extractall(r"(#\S+)")
    freqs = hashtags[0].value_counts().reset_index()
    freqs.columns = ["hashtag", "count"]
    freqs2= freqs.head(10)
    display(freqs2)
```
